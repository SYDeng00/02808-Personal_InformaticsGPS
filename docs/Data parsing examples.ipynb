```python
import json
from datetime import datetime

def calculate_duration(start, end):
    start_dt = datetime.fromisoformat(start[:-1])
    end_dt = datetime.fromisoformat(end[:-1])
    return (end_dt - start_dt).total_seconds()

def sum_durations_from_file(file_path):
    with open(file_path, 'r') as file:
        data = json.load(file)

    durations = {}

    for timeline_obj in data['timelineObjects']:
        for key, value in timeline_obj.items():
            if "duration" in value:
                duration = calculate_duration(value['duration']['startTimestamp'], value['duration']['endTimestamp'])
                if key not in durations:
                    durations[key] = duration
                else:
                    durations[key] += duration

    for location, total_duration in durations.items():
        print(f"{location}: {total_duration} seconds")

# File path to your JSON file
file_path = 'data/2023_JUNE.json'

# Call the function with the path to your JSON file
sum_durations_from_file(file_path)
```
